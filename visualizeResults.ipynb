```python
## show predictions & get F1

import os
import h5py
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
from sklearn.metrics import f1_score, precision_recall_fscore_support, confusion_matrix
import statistics
from skimage.transform import resize

directory = '/geomatics/gpuserver-0/scuypers/CCT_output_L4S/FL_weakly_320_selection_resume/'
#directoryGT = '/srvgentjkd98p2/K/Projects/Satellite_Photogrammetry/LandSlide/ValidationData/mask_best/'
directoryGT = '/srvgentjkd98p2/K/Projects/Satellite_Photogrammetry/LandSlide/TestData/mask_best/'
#directoryGT = '/srvgentjkd98p2/K/Projects/Satellite_Photogrammetry/LandSlide/TrainData/mask/'

tn, fp, fn, tp = 0, 0, 0, 0
aantal = 30 # number of images to show
index = 0
for filename in os.listdir(directory):
    if filename.endswith(".h5") and index < aantal:
        index += 1
        print(filename)

        f = h5py.File(directory+filename, 'r')
        dset = f['mask']

        # mask
        data = np.array(dset[:,:])
        #print(data.shape)
        unique = np.unique(data)
        print(unique)
        plt.figure(figsize=(5,5))
        plt.imshow(data)
        plt.title('prediction')
        plt.show()
                
        # GT of mask_best
        fGT = h5py.File(directoryGT+filename,'r')
        dsetGT = fGT['mask']
        
        # original size GT
        dataGT = np.array(dsetGT[:,:])
        uniqueGT = np.unique(dataGT)
        #print(uniqueGT)
        plt.figure(figsize=(5,5))
        plt.imshow(dataGT)
        plt.title('GT')
        plt.show()

        
        result = data.flatten()
        resultGT = dataGT.flatten()
        F1_score = f1_score(resultGT, result, pos_label=1, average='binary')
        print(F1_score)
        p, r, f, s= precision_recall_fscore_support(resultGT, result, pos_label=1)
        print(p,r,f,s)
        if len(uniqueGT) == 1 and len(unique) == 1:
            tn1 = confusion_matrix(resultGT, result).ravel()
            print(tn1)
            fp1, fn1, tp1 = 0, 0, 0
        else:
            tn1, fp1, fn1, tp1 = confusion_matrix(resultGT, result).ravel()
        print( tn1, fp1, fn1, tp1)
        tn += tn1
        fp += fp1
        fn += fn1
        tp += tp1
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)
print('precision:',precision,'recall:', recall,'f1:', f1)
```
